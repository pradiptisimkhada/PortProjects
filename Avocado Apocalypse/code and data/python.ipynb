DATA ANALYSIS

The metrics that interested me the most were: the date, average price, the total volume, and the volume per PLU codes which indicates the size of the avocadoes ( plu4046 : Hass Small ,
                                            plu4225 : Hass Large ,
                                            plu4770 : Hass Extra Large)


The general volume sold as well as the sales for each variety throughout the years.

```python
# Creating a line plot using Seaborn
sns.set(style="darkgrid") 
plt.figure(figsize=(8, 6)) 

# Plot the line chart
sns.lineplot(data=avoc, x='year', y='TotalVolume', marker='o', label='Total Volume', errorbar=None)
sns.lineplot(data=avoc, x='year', y='plu4046', marker='o', label='Small(plu4046)', errorbar=None)
sns.lineplot(data=avoc, x='year', y='plu4225', marker='o', label='Large(plu4225)',  errorbar=None)
sns.lineplot(data=avoc, x='year', y='plu4770', marker='o', label='Extra Large(plu4770)', errorbar=None)

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Sold in Billions(Volume)")
plt.title("Sales of each variety of Avocadoes through out the years")

plt.legend()
plt.show()
```

It looks like the PLU4225(Hass Large size) and PLU4046 (Hass Small size) are chasing each other until 2018 when PLU4046 became the most sold.

Looking at the price, to see if it followed the trend of the volume sold:

```python
plt.figure(figsize=(10, 6))

# Plot AveragePrice
sns.lineplot(data=avoc, x="year", y="AveragePrice", label="Average Price",errorbar=None)

# Plot TotalVolume on a secondary y-axis
ax2 = plt.gca().twinx()
sns.lineplot(data=avoc, x="year", y="TotalVolume", label="Total Volume", ax=ax2, color='orange', errorbar=None)

plt.xlabel("Year")
plt.ylabel("Average Price in $")
ax2.set_ylabel("Total Volume")

plt.title("Price and Sales Evolutions")

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

From 2019 to 2020, the average price went down from $1.525 to $1.385 but the sales went up significantly at the same time. From 2020 to 2021, the price continued to drop but the consumption dropped. So we can conclude  the price alone is not sufficient enough to predict the sales, the quality of the product, the trends, the taste of people, in 2019 the COVID appeared so that surely played an effect on the sales of avocados.

```python
#Plotting a line chart to see sales of each variety of avocadoes
sns.lineplot(data=avoc, x="year", y="plu4046", label="PLU 4046",errorbar=None)
sns.lineplot(data=avoc, x="year", y="plu4225", label="PLU 4225",errorbar=None)
sns.lineplot(data=avoc, x="year", y="plu4770", label="PLU 4770",errorbar=None)


plt.xlabel("Year")
plt.ylabel("Cumulative Sales")
plt.title("Sales Comparison of varieties of Avocado")


plt.legend()

plt.show()
```

Clearly plu4046(Small) is the most sold in 2021.

```python
avoc['region'].unique()
```

 Looking at the areas we've got the data per city, as well as per region so we'll compare both.

```python
avoc_region = avoc[avoc['region'].isin(['West','California','SouthCentral','Northeast','Southeast','GreatLakes','MidSouth','Plains'])].sort_values(by="TotalVolume", ascending=False)
avoc_region
```

```python
avoc_city = avoc[avoc['region'].isin(['LosAngeles','NewYork','DallasFtWorth','PhoenixTucson','Houston','Denver','BaltimoreWashington','WestTexNewMexico','SanFrancisco','Chicago'])].sort_values(by="TotalVolume", ascending=False)
avoc_city
```

```python
#For the region
sns.barplot(data=avoc_region, x="region", y="TotalVolume",errorbar=None)

plt.xlabel("Region",)
plt.ylabel("Total Volume Sold")
plt.title("Total Volume Sold in Selected Regions")
plt.xticks(rotation=45)

plt.show()

#For the city
sns.barplot(data=avoc_city, x="region", y="TotalVolume",errorbar=None)

plt.xlabel("City",)
plt.ylabel("Total Volume Sold")
plt.title("Total Volume Sold in Selected Cities")
plt.xticks(rotation=45)

plt.show()
```

Looking at the data of the specific city LosAngeles where vegan and vegetarian lifestyle is quite popular and seemingly biggest avocado consuming city to see if it reflects the data of whole USA.

```python
avoc_la = avoc[avoc['region']=='LosAngeles']
avoc_la
```

```python
sns.lineplot(data=avoc_la, x="year", y="TotalVolume", label="Total Volume",errorbar=None)
sns.lineplot(data=avoc_la, x="year", y="plu4046", label="PLU 4046",errorbar=None)
sns.lineplot(data=avoc_la, x="year", y="plu4225", label="PLU 4225",errorbar=None)
sns.lineplot(data=avoc_la, x="year", y="plu4770", label="PLU 4770",errorbar=None)


plt.xlabel("Year")
plt.ylabel("Sales")
plt.title("Units of Avocadoes sold in LA")


plt.legend()

plt.show()
```

We can see that the sales of the wholes country vastly diifers from the sales in LA

Looking at the top 10 cities in terms of volume

```python
city_volume = avoc_city.groupby(["region"])["TotalVolume"].sum()

# Get top 10 cities in terms of volume
top_10_cities = city_volume.nlargest(10).sort_values(ascending=False)
top_10_cities_df = top_10_cities.reset_index(name="TotalVolume")
top_10_cities_df
```

By checking the top 10 cities in terms of volume from 2015 to 2021, we can clearly see that Los Angeles and Dallas are the two majors contributors to the sales of avocados in the U.S.
